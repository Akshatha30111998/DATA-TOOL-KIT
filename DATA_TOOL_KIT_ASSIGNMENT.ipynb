2.How does broadcasting work in Numpy?
- Broadcasting is a powerful feature in NumPy that allows you to perform operations on arrays with different shapes and sizes.
- Broadcasting is the process of aligning arrays with different shapes and sizes so that they can be operated on element-wise.



```python

import numpy as np

a = np.array([1, 2, 3])
b = 2

print(a + b)  # Output: [3, 4, 5]
```

- 1. Compatible shapes: The arrays must have compatible shapes. Two arrays are compatible if their shapes can be made to match by adding new axes (dimensions) to one or both arrays.
- 2. Size 1 dimensions: When operating on two arrays, if one array has a size 1 dimension, it will be broadcasted to match the size of the corresponding dimension in the other array.
- 3. Singleton dimensions: A singleton dimension is a dimension of size 1. When broadcasting, singleton dimensions are ignored.

4. Explain the use of the groupby() method in Pandas?
- The groupby() method in Pandas is a powerful tool for grouping data by one or more columns and performing aggregation operations on the grouped data.
 - Parameters

- by: The column(s) to group by. Can be a string, a list of strings, or a pandas Series.

6.What are the differences between NumPy arrays and Python lists?
 - NumPy arrays and Python lists are both data structures used to store collections of data. However, they have distinct differences in terms of their characteristics, advantages, and use cases.

Differences:

- 1. Homogeneous vs. Heterogeneous:
    - NumPy arrays: Homogeneous, meaning all elements must be of the same data type.
    - Python lists: Heterogeneous, meaning elements can be of different data types.
- 2. Memory Layout:
    - NumPy arrays: Stored in contiguous blocks of memory, allowing for efficient caching and vectorized operations.
    - Python lists: Stored as dynamic arrays, with each element being a separate object, leading to slower performance.
- 3. Performance:
    - NumPy arrays: Optimized for numerical computations, providing faster performance and vectorized operations.
    - Python lists: General-purpose data structure, with slower performance compared to NumPy arrays.
-4. Multidimensionality:
    - NumPy arrays: Native support for multidimensional arrays, making it easier to work with matrices and higher-dimensional data.
    - Python lists: Limited support for multidimensional data, requiring workarounds like nested lists.
-5. Operations:
    - NumPy arrays: Support for element-wise operations, matrix multiplication, and other advanced mathematical operations.
    - Python lists: Limited support for element-wise operations, requiring explicit loops or list comprehensions.
-6. Size Limitations:
    - NumPy arrays: Limited by available memory, but can handle large datasets.
    - Python lists: Limited by available memory, but can become slow and inefficient for very large datasets.
-7. Data Type Conversion:
    - NumPy arrays: Automatic data type conversion for compatible operations.
    - Python lists: No automatic data type conversion, requiring explicit conversions.

Vectorized operations are a key feature of NumPy and provide several benefits, including:

17.Explain the difference between apply() and map() in Pandas?
- In Pandas, apply() and map() are two popular functions used to perform operations on data. While they share some similarities, there are key differences between them:

- Apply()
- apply() is a more general-purpose function that applies a function to each row or column of a DataFrame or Series.
- It can handle more complex operations, such as applying a function to multiple columns or rows.
- apply() can return a transformed DataFrame or Series, or even a scalar value.
- It's often used for data cleaning, filtering, or transforming data.













```python

import pandas as pd

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})

def square(x):
    return x**2

result = df['A'].apply(square)
print(result)
```


- Map()
- map() is a function that applies a mapping (i.e., a dictionary or a function) to each element of a Series.
- It's primarily used for replacing values in a Series based on a mapping.
- map() returns a new Series with the transformed values.
- It's often used for data cleaning, encoding categorical variables, or replacing missing values.


```python

import pandas as pd

s = pd.Series(['a', 'b', 'c'])

mapping = {'a': 1, 'b': 2, 'c': 3}

result = s.map(mapping)
print(result)
```

21.Why is NumPy’s array slicing faster than Python’s list slicing?
- NumPy's array slicing is faster than Python's list slicing for several reasons:

- Contiguous Memory Allocation
NumPy arrays store elements in contiguous blocks of memory. This allows for efficient slicing, as NumPy can simply return a new array object that points to the same memory block, but with adjusted offsets and shapes.

- Vectorized Operations
NumPy's slicing operations are implemented in C, which provides a significant performance boost compared to Python's list slicing, which is implemented in Python.

- Reduced Memory Allocation
When slicing a NumPy array, a new array object is created, but it does not allocate new memory for the sliced elements. Instead, it shares the same memory block as the original array. This reduces memory allocation overhead.

- Cache Locality
NumPy arrays are designed to optimize cache locality, which means that elements that are close together in memory are also close together in the cache. This leads to faster access times and improved performance.

- List Slicing Overhead
Python's list slicing, on the other hand, involves creating a new list object and copying the sliced elements into the new list. This involves additional memory allocation and copying overhead, which can be slower than NumPy's array slicing.

```python
#6. Create a DataFrame and add a new column based on an existing column?
# Import the Pandas library
import pandas as pd

# Create a DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Dave'],
    'Age': [25, 30, 35, 40]
}
df = pd.DataFrame(data)

# Print the original DataFrame
print("Original DataFrame:")
print(df)

# Add a new column 'Adult' based on the 'Age' column
df['Adult'] = df['Age'].apply(lambda x: 'Yes' if x >= 18 else 'No')

# Print the updated DataFrame
print("\nUpdated DataFrame:")
print(df)
```

```python
#8.Create a line plot with multiple lines using Matplotlib?

# Import the Matplotlib library
import matplotlib.pyplot as plt

# Import the NumPy library for generating data
import numpy as np

# Generate data for the plot
x = np.linspace(0, 10, 100)

# Define multiple lines
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(2*x)

# Create the plot
plt.figure(figsize=(8, 6))

# Plot the lines
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.plot(x, y3, label='sin(2x)')

# Add title and labels
plt.title('Multiple Lines Plot')
plt.xlabel('x')
plt.ylabel('y')

# Add legend
plt.legend()

# Display the plot
plt.show()
```

```python
#9.A Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold?
# Import the Pandas library
import pandas as pd

# Generate a sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Dave', 'Emily'],
    'Age': [25, 30, 35, 40, 45],
    'Score': [90, 85, 95, 80, 92]
}
df = pd.DataFrame(data)

# Print the original DataFrame
print("Original DataFrame:")
print(df)

# Define the threshold
threshold = 90

# Filter rows where the 'Score' is greater than the threshold
filtered_df = df[df['Score'] > threshold]

# Print the filtered DataFrame
print("\nFiltered DataFrame (Score > {}):".format(threshold))
print(filtered_df)
```

```python
#10.Create a histogram using Seaborn to visualize a distribution?

# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Generate a random sample from a normal distribution
np.random.seed(0)
data = np.random.randn(1000)

# Create a histogram using Seaborn
plt.figure(figsize=(8, 6))
sns.histplot(data, bins=30, kde=True)

# Add title and labels
plt.title('Histogram of Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Display the plot
plt.show()

```

```python
#11.Perform matrix multiplication using NumPy?
# Import the NumPy library
import numpy as np

# Define two matrices
matrix_a = np.array([[1, 2], [3, 4]])
matrix_b = np.array([[5, 6], [7, 8]])

# Print the matrices
print("Matrix A:")
print(matrix_a)
print("\nMatrix B:")
print(matrix_b)

# Perform matrix multiplication
result = np.matmul(matrix_a, matrix_b)

# Print the result
print("\nResult of Matrix Multiplication:")
print(result)

```

```python
#12.Use Pandas to load a CSV file and display its first 5 rows?

# Import the Pandas library
import pandas as pd

# Load the CSV file
def load_csv(file_path):
    try:
        df = pd.read_csv(file_path)
        return df
    except FileNotFoundError:
        print("File not found. Please check the file path.")
        return None
    except pd.errors.EmptyDataError:
        print("File is empty. Please check the file contents.")
        return None
    except pd.errors.ParserError:
        print("Error parsing the file. Please check the file format.")
        return None

# Display the first 5 rows of the DataFrame
def display_first_rows(df, num_rows=5):
    if df is not None:
        print(df.head(num_rows))

# Usage example
file_path = 'data.csv'  # Replace with your CSV file path
df = load_csv(file_path)
display_first_rows(df)
```

```python
#13.Create a 3D scatter plot using Plotly.
# Import the Plotly library
import plotly.graph_objects as go
import numpy as np

# Generate random data for the 3D scatter plot
np.random.seed(0)
x = np.random.randn(100)
y = np.random.randn(100)
z = np.random.randn(100)

# Create the 3D scatter plot
fig = go.Figure(data=[go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='markers',
    marker=dict(
        size=5,
        color=z,  # Color the markers based on the z-values
        colorscale='Viridis',
        opacity=0.8
    )
)])

# Customize the plot
fig.update_layout(
    title='3D Scatter Plot',
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z'
    )
)

# Display the plot
fig.show()

```
